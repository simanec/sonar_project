### Short description

The dataset contains 111 patterns obtained by bouncing sonar
signals off a metal cylinder at various angles and under various
conditions. The file "sonar.rocks" contains 97 patterns obtained from
rocks under similar conditions. The transmitted sonar signal is a
frequency-modulated chirp, rising in frequency. The data set contains
signals obtained from a variety of different aspect angles, spanning 90
degrees for the cylinder and 180 degrees for the rock.

Each pattern is a set of 60 numbers in the range 0.0 to 1.0. Each number
represents the energy within a particular frequency band, integrated over
a certain period of time. The integration aperture for higher frequencies
occur later in time, since these frequencies are transmitted later during
the chirp.

The label associated with each record contains the letter "R" if the object
is a rock and "M" if it is a mine (metal cylinder). The numbers in the
labels are in increasing order of aspect angle, but they do not encode the
angle directly


The task can be defined as binary classification with positive class 'Mines' because this type of objects can be concidered as more "desirable".


link: https://www.kaggle.com/datasets/rupakroy/sonarcsv

## Preparation of data

The following function will be used to transform mines class labels 'M' to 1 (positive class) and Rocks class labels 'R' to 0 (negative class).

```python
def custom_transform(Y_array):
    N = len(Y_array)
    Y_tr =[]
    for k in range(N):
        if Y_array[k] == 'M':
    #         mine is a "positive" class
            Y_tr.append(1)
        else:
    #         the rest (=Rocks) is negative class
            Y_tr.append(0)
    Y_tr = np.array(Y_tr)
    return Y_tr
```

```python
array = data.values
X = array[:,0:60] #features
Y = array[:,60] #target
```

Splitting the original dataset into test set (20% of original set),
validation set (20% of the rest dataset = 16% of original dataset) and training set (60% of original set)

Together validation and trainig set are named 'training_0' set (80% of original)

```python
seed = 7
test_size = 0.2
validation_size = 0.2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# all -> test + train_0
X_train_0, X_test, Y_train_0, Y_test = train_test_split(X, Y, test_size=test_size, random_state=seed)
```

```python
#train_0 -> train + validation
# validation_size = validation_size/(1-test_size)
X_train, X_validation, Y_train, Y_validation = train_test_split(X_train_0, Y_train_0,
                                                                test_size=validation_size, random_state=seed)
```

Binarazing targets for all of the subsets (M->1, R->0):

```python
custom_transform(np.array(['M', 'R', 'R']))
```
